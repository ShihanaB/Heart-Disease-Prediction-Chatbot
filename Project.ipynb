```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("fedesoriano/heart-failure-prediction")

print("Path to dataset files:", path)
```

```python
import os

# List files in the downloaded directory
files = os.listdir(path)
print(files)
```

```python
import pandas as pd

# Construct the full path to the CSV file
csv_file_path = os.path.join(path, 'heart.csv')

df = pd.read_csv(csv_file_path)
df.head()
```

```python
hf = h2o.H2OFrame(df)
target = "HeartDisease"
features = hf.columns
features.remove(target)
```

```python
hf.describe()
```

```python
# Convert the target column to a factor for binary classification
hf['HeartDisease'] = hf['HeartDisease'].asfactor()
```

```python
train, test = hf.split_frame(ratios=[0.8], seed=1234)
```

```python
from h2o.automl import H2OAutoML

aml = H2OAutoML(max_runtime_secs=120, seed=42)
aml.train(x=features, y=target, training_frame=train)
```

```python
perf = aml.leader.model_performance(test)
perf
```

```python
preds = aml.leader.predict(test)
preds.head()
```

```python
model_path = h2o.save_model(model=aml.leader, path="./", force=True)
print("Model saved to:", model_path)
```
